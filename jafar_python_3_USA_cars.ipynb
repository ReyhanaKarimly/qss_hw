__Disclaimer: This dataset contains new released car models between 1999 and 2008.__  
Variable dictionary is listed below:  
_Manufacturer:_ manufacturer name  
_Model:_ model name  
_Displ:_ engine displacement, in litres  
_Year:_ year of manufacture  
_Cyl:_ number of cylinders  
_Trans:_ type of transmission  
_Drv:_ the type of drive train, where f = front-wheel drive, r = rear wheel drive, 4 = 4wd  
_Cty:_ city miles per gallon  
_Hwy:_ highway miles per gallon  
_Fl:_ fuel type  
_Class:_ "type" of car   
Please, answer following questions:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
cars = pd.read_excel('cars.xlsx')
cars
```

```python
# 1. Replace NaNs with mean for highway miles per gallon variable.

cars['hwy'] = cars['hwy'].fillna(cars['hwy'].mean())
```

```python
# 2. Delete observations that have NaN values.

cars = cars.dropna()
```

```python
# 3. Find correlation between variables. Which variables have highest correlation?

corr_matrix = cars.corr()
corr_matrix
#ANSWER: cty and hwy have the highest correlation (0.954192)
```

```python
# 4. Show relationship between highway and city miles per gallon.
# 4.a. Use scatter in matplotlib.
plt.scatter(x=cars['hwy'], y=cars['cty'])
```

4.b. Interpret how are variables correlated according to scatter.  
ANSWER: There is positive correlation, but it costs more gallons per mile in city.

```python
# 5. Show distributions and scatters between all variables. While hue equals to type of drive train.
# Which variables have the lowest and the highest correlation? Find according to scatterplot.

fig, axs = plt.subplots(5, 2, figsize=(10,20))
sns.histplot(cars, x='manufacturer', hue='drv', ax=axs[0][0])
sns.histplot(cars, x='model', hue='drv', ax=axs[0][1])
sns.histplot(cars, x='displ', hue='drv', ax=axs[1][0])
sns.histplot(cars, x='year', hue='drv', ax=axs[1][1])
sns.histplot(cars, x='cyl', hue='drv', ax=axs[2][0])
sns.histplot(cars, x='trans', hue='drv', ax=axs[2][1])
sns.histplot(cars, x='cty', hue='drv', ax=axs[3][0])
sns.histplot(cars, x='hwy', hue='drv', ax=axs[3][1])
sns.histplot(cars, x='fl', hue='drv', ax=axs[4][0])
sns.histplot(cars, x='class', hue='drv', ax=axs[4][1])
```

```python
# 5. (continued)

sns.pairplot(data=cars, hue='drv') # only numerical variables

# ANSWER: cty vs. year, and hwy vs. year have low correlation.
```

```python
# 6. Which type of car is most frequent in dataset? Show by using countplot.

sns.countplot(x=cars['class'])

#ANSWER: suv is the most frequent type of car.
```

```python
# 7. Display number of cylinders for each drive train in bar chart. Which drive train is the most frequent?

sns.barplot(cars['drv'], cars['cyl'])

# ANSWER: rear wheel drive is the most frequent.
```

```python
# 8. Visualize engine displacement by each class, using boxplot. Do the same thing in violinplot.

fig, axs = plt.subplots(1, 2, figsize=(16,6))
sns.boxplot(data=cars, x='class', y='displ', ax=axs[0])
sns.violinplot(data=cars, x='class', y='displ', ax=axs[1])
```
