```python
import pandas as pd
import seaborn as sns
```

```python
# 1) Import dataset of avocado.

avocado = pd.read_csv('avocado.csv')
avocado
```

```python
# 2) Remove Unnamed: 0 variable.

avocado = avocado.drop('Unnamed: 0', axis=1)
```

```python
# 3) List number of null values for each variable.

null_lists = []

for variable in avocado.columns:
    null_values = avocado[variable][avocado[variable].isna()]
    null_lists.append(null_values)
    
print(null_lists) #there are no null values
```

```python
# 4) Illustrate relationship between Total Bags and Small Bags using scatter plot.

sns.scatterplot(x=avocado['Total Bags'], y=avocado['Small Bags'])
```

```python
# 5) Construct Price distribution.

sns.displot(x=avocado['AveragePrice'])
```

```python
# 5.a) Which type of Avocados are expensive?

sns.displot(avocado, x='AveragePrice', hue='type') # Organic avocados are more expensive.
```

```python
# 6) Find relationship between all variables using scatterplot.
# I removed "date" column, as it is duplicated by year

sns.pairplot(avocado, vars=['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags', 'type', 'region', 'year'])
```

```python
# 7) Find relationship between large and small bags using scatterplot.

sns.scatterplot(data=avocado, x='Large Bags', y='Small Bags')
```
